```python
import pandas as pd
import numpy as np
import random

# set random seed for reproductivity
np.random.seed(42)

#generate data
num_rows = 100
data={
    'User ID' : range(1, num_rows + 1),
    "Gender": [random.choice(['Male', 'Female']) for _ in range(num_rows)],
    'Age' : [random.randint(18 , 65) for _ in range(num_rows)],
    'Estimated Salary' : np.random.randint(20000, 200000, num_rows),
    'Purchased' : np.random.randint(0, 2, num_rows)
}

#Create dataframe
df = pd.DataFrame(data)
display(df)
```

```python
df.to_csv('Social_Network_Ads.csv', index=False)
```

```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Social_Network_Ads.csv')
dataset
```

```python
x = dataset.iloc[:, [2,3]].values
y = dataset.iloc[:,4].values
display(x)
display(y)
```

```python
# Splitting the dataset into training and test sets
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=0)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
st_x = StandardScaler()
x_train = st_x.fit_transform(x_train)
x_test = st_x.transform(x_test)
```

```python
x_test
```

Fitting K-NN classifier to the the Training data: Now we will fit the K-NN classifier to the training data. To do this we will import the KNeighborsClassifier class of the Sklearn Neigh

```python
#Fitting KNN classifier to the training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```
