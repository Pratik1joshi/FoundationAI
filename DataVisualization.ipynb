```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# Line plot
x = np.linspace(0,10,100)
y = np.tan(x)
```

```python
# print(x)
# print("_________________")
# print(y)
```

```python
plt.plot(x,y,label = "Sinusoidal wave")
plt.title("Line graph")
plt.xlabel("X-axis ----->")
plt.ylabel("Y-axis ----->")
plt.legend()
plt.show()
```

```python
#Barchart
categories = ["A", "B", "C", "D"]
values = [100, 150, 120, 80]
plt.bar(categories, values)
colors = ["red", "green", "blue", "yellow"]
plt.bar(categories, values, color=colors)
plt.title("Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Histogram and KDE plot
data = np.random.randn(1000)
sns.histplot(data, kde=True, color = "red")
plt.title("Histogram with KDE")
plt.show()
```

```python
#Heat map
matrix = np.random.rand(5,5)
sns.heatmap(matrix, annot=True, cmap="YlGnBu",fmt=".2f")
plt.title("Heatmap")
plt.show()
```

```python
#pairplot
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue = "species", markers = "D")
plt.show()
```
